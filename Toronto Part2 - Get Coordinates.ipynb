```python
import pandas as pd
df = pd.read_csv('Toronto.csv', index_col=0)
df_co = pd.read_csv('Geospatial_Coordinates.csv', index_col=0)
```

```python
Latitude = list()
Longitude = list()
for postalcode in df['PostalCode']:
    Latitude.append(df_co.loc[postalcode, 'Latitude'])
    Longitude.append(df_co.loc[postalcode, 'Longitude'])
```

```python
df['Latitude'] = Latitude
df['Longitude'] = Longitude
```

```python
df.to_csv('Toronto2.csv')
df.head(12)
```
